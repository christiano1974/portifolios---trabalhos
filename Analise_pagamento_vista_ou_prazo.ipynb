[8] (valor 2,0 pontos) Um apartamento pode ser comprado à vista por $ 320.000,00 ou pagando 20% e duas prestações. A primeira para 3 meses após a venda e a segunda para sete meses após a venda. Considere a taxa de 3,5% ao mês com capitalização a juros compostos. Determine:
(A) Calcule o valor de cada prestação. Justifique sua resposta.
(B) Se a taxa de juros vigentes no mercado para aplicações for de 2% a.m. a juros compostos, então qual será a melhor opção de compra? Justifique sua resposta.

```python
import numpy_financial as npf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# solução do item (A) do exercício 8 - primeira prova 2-2023
P = 256000*(1.035**(-3) + 1.035**(-7))**(-1)
P = round(P,2)
print("Valor de cada prestação: R$",P)
```

```python
# solução do item (B) do exercício 8 - primeira prova 2-2023
FC = [0,0,0,P,0,0,0,P]
aplicacao = round(npf.npv(0.02,FC),2)
print("Valor presente da aplicação: R$",aplicacao)
entrada = round(0.20*320000.00,2)
print("Custo da pagamento a prazo: R$",entrada + aplicacao)
print("Conlusão: Com taxa de retorno de 2 % a.m. o pagamento a prazo não deve ser realizado.")
```

```python
# Gráfico comparativo entre pagamento à vista e pagamento a prazo
taxas_retorno = np.arange(0.0,0.06,0.01)
custo_pagamentos_prazo = [npf.npv(k,FC) + entrada for k in taxas_retorno]
vetor = np.zeros(len(taxas_retorno))
for k in range(len(vetor)):
    vetor[k] = 320000
plt.figure(figsize = (10,5))
plt.plot(taxas_retorno,custo_pagamentos_prazo,marker = '*',linewidth = 2)
plt.plot(taxas_retorno,vetor,'r--')
plt.title('Pagamento a prazo com taxa de aplicação (retorno) i',fontsize = 12)
plt.xlabel('Taxas de retorno i',fontsize = 10)
plt.ylabel('Valor do pagamento a prazo',fontsize = 10)
plt.xlim = [0.0,0.05]
plt.grid()
plt.legend(labels = ['custo do pagamento a prazo','custo do pagamento à vista'],loc ='upper center')
plt.show()
```
