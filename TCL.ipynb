# Segue abaixo um exemplo considerando a distribuição binominal

```python
from scipy.stats import binom,norm
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Distribuição Binomial com n = 100 e p = 0.50 (simulação do lançamento de uma moeda honesta)
x_valores = list(range(101))
mean, var = binom.stats(100,0.5)
dist_1 = [binom.pmf(x,100,0.5) for x in x_valores]
```

```python
# Gráfico da distribuição binomial
plt.figure(figsize = (10,5))
plt.bar(x_valores,dist_1)
plt.title('Distribuição binomial Bin(100,0.5)',fontsize = 12)
plt.xlabel('valores de x',fontsize = 10)
plt.ylabel('probabilidade',fontsize = 10)
plt.show()
```

```python
# Parâmetros da distribuição Binomial
print('média = ' + str(mean))
print('variância = ' + str(var))
```

```python
# Gráfico da distribuição normal com média n.p e variância n.p.(1-p)
plt.figure(figsize = (10,5))
plt.plot(x_valores,norm.pdf(x_valores,50,5),color = 'red')
plt.title('Distribuição normal N(50,25)',fontsize = 12)
plt.xlabel('valores de x',fontsize = 10)
plt.ylabel('probabilidade',fontsize = 10)
plt.show()
```

```python
# Teorema Central do Limite
plt.figure(figsize = (10,5))
plt.bar(x_valores,dist_1)
plt.plot(x_valores,norm.pdf(x_valores,50,5),color = 'red',linewidth = 2)
plt.title('Teorema do Limite Central - TLC (n = 100)')
plt.xlabel('valores de x',fontsize = 10)
plt.ylabel('probabilidade',fontsize = 10)
plt.legend(labels = ['dist. binomial','dist. normal'],loc = "upper left")
plt.show()
```
